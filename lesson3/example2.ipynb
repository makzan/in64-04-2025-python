```python
import requests
import pandas as pd
import time
```

```python
dfs = []
for page_index in range(1,4):
    url=f"https://mo.centanet.com/Transaction/Index?isFrame=&FArea=&Name=&FileType=%E4%BD%8F%E5%AE%85&PropType=S&TitleName=2025%E5%B9%B4&BeginTxDate=2025-01-01&EndTxDate=2025-5-8&DateValue=2025&DataType=Y&PageIndex={page_index}"
    res = requests.get(url)
    tables = pd.read_html(res.text)
    dfs.append(tables[0])
    time.sleep(0.5)
len(dfs)
```

```python
df = pd.concat(dfs)
df
```

```python
df.dtypes
```

```python
df["成交價"] = df["成交價"].str.replace(",","").str.replace("$","").str.replace("萬","").astype(int)
```

```python
df["建築面積(呎)"] = df["建築面積(呎)"].str.replace(",", "").str.replace("呎","").astype(int)
```

```python
len(df)
```

```python
result_len = df.groupby("片區")["成交價"].agg(len)
result_min = df.groupby("片區")["成交價"].agg(min)
result_max = df.groupby("片區")["成交價"].agg(max)
df_result = pd.concat([result_len, result_min, result_max], axis=1)
df_result.columns = ["Count", "成交價 min", "成交價 max"]
df_result
```
